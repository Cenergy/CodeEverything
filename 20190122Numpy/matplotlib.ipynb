```python
import matplotlib as mpl
import  matplotlib.pyplot as plt
```

```python
import numpy as np

a=np.linspace(0,10,100)
a
```

```python
b=np.sin(a)
b
```

```python
plt.plot(a,b)
plt.show()
```

```python
c=np.cos(a)
plt.plot(a,b)
plt.plot(a,c)
plt.show()
```

```python
plt.plot(a,b,linestyle=":")
plt.plot(a,c,color='r')
plt.xlim(-5,15)
plt.show()
```

```python
plt.plot(a,b,linestyle=":",label="sin(x)")
plt.plot(a,c,color='r',label="cos(x)")
plt.axis([-1,11,-2,2])
plt.ylabel("y value")
plt.xlabel("x axis")
plt.legend()
plt.title("Hello matplotlib")
plt.show()
```

##  scatter point

```python
plt.scatter(a,b)
plt.show()
```
